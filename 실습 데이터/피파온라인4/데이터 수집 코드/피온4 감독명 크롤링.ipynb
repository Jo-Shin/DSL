```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
from random import uniform

def random_sleep():
    time.sleep(uniform(1.5, 2.5))
```

```python
random_sleep()
```

```python
PATH = "/Users/dongwook/chromedriver"
driver = webdriver.Chrome(PATH)

# 피온 4 사이트 접속
driver.get("http://fifaonline4.nexon.com/main/index")
random_sleep()


# 공식 경기 랭킹
search = driver.find_element_by_xpath('//*[@id="header"]/div[1]/div/ul/li[6]/a/span')
search.click()
random_sleep()

ranking = driver.find_element_by_xpath('//*[@id="middle"]/div/div/div[1]/div[2]/ul/li[5]/a')
ranking.click()
random_sleep()


# 감독명 가져오기
names = []
for k in range(49):
    
    for j in range(1, 11):

        elements = driver.find_elements_by_class_name("name.profile_pointer")
        random_sleep()
        for i in range(20):
            ele = elements[i].text
            names.append(ele)
        
        next_page = driver.find_element_by_xpath('//*[@id="middle"]/div/div/div[3]/div[2]/div/ul/li[{}]/a'.format(j))
        next_page.click()
        random_sleep()
    
    next_section = driver.find_element_by_class_name('btn_next_list.ajaxNav')
    next_section.click()
    random_sleep()
    print("현재까지 크롤링된 감독명 개수: {}".format(len(names)))
```
